```python
# import statements
from time import sleep
import time
import datetime
import datetime as dt
from datetime import datetime, timezone
from json import dumps
from kafka import KafkaProducer
import random
import pandas as pd

# Publising data
def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        tim = time.time()
#         print('Message published successfully. At {}'.format(tim) + ' Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
        # connection fo kafka
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'process_producer'
    
    print('Publishing records..')
    # Kafka Producer
    producer = connect_kafka_producer()
    # Reading in data
    data = pd.read_csv('Streaming_Linux_process.csv')
    
    # Splitting data based on machines
    M_4 = data.loc[data.machine == 4].reset_index(drop=True)
    M_5 = data.loc[data.machine == 5].reset_index(drop=True)
    M_6 = data.loc[data.machine == 6].reset_index(drop=True)
    M_7 = data.loc[data.machine == 7].reset_index(drop=True)
    M_8 = data.loc[data.machine == 8].reset_index(drop=True)
    
    
    while True:
        # random number generator for machines
        k4 = random.randint(10,50)
        k5 = random.randint(10,50)
        k6 = random.randint(10,50)
        k7 = random.randint(10,50)
        k8 = random.randint(10,50)
        # timestamp UTC based
        machine_timestamp = int(datetime.now().replace(tzinfo = timezone.utc).timestamp())
        M_4['ts'] = machine_timestamp
        M_5['ts'] = machine_timestamp
        M_6['ts'] = machine_timestamp
        M_7['ts'] = machine_timestamp
        M_8['ts'] = machine_timestamp

        to_send = []
        
        # Checking conditions if data is less than 50 in dataframe for all machines
        if len(M_4) < k4:
            if len(M_4) < 10:
                M_4 = data.loc[data.machine == 4].reset_index(drop=True)
            M_4_Data = M_4[:].to_dict(orient = 'record')
            M_4.drop(M_4.index, inplace=True)
            M_4 = data.loc[data.machine == 4].reset_index(drop=True)
        else:
            M_4_Data = M_4[:k4].to_dict(orient = 'record')
            M_4.drop(list(range(0,k4)), inplace=True)
            M_4.reset_index(drop=True, inplace = True)
            
        if len(M_5) < k5:
            if len(M_5) < 10:
                M_5 = data.loc[data.machine == 5].reset_index(drop=True)
            M_5_Data = M_5[:].to_dict(orient = 'record')
            M_5.drop(M_5.index, inplace=True)
            M_5 = data.loc[data.machine == 5].reset_index(drop=True)
        else:
            M_5_Data = M_5[:k5].to_dict(orient = 'record')
            M_5.drop(list(range(0,k5)), inplace=True)
            M_5.reset_index(drop=True, inplace = True)

            
        if len(M_6) < k6:
            if len(M_6) < 10:
                M_6 = data.loc[data.machine == 6].reset_index(drop=True)
            M_6_Data = M_6[:].to_dict(orient = 'record')
            M_6.drop(M_6.index, inplace=True)
            M_6 = data.loc[data.machine == 6].reset_index(drop=True)
        else:
            M_6_Data = M_6[:k6].to_dict(orient = 'record')
            M_6.drop(list(range(0,k6)), inplace=True)
            M_6.reset_index(drop=True, inplace = True)

            
        if len(M_7) < k7:
            if len(M_7) < 10:
                M_7 = data.loc[data.machine == 7].reset_index(drop=True)
            M_7_Data = M_7[:].to_dict(orient = 'record')
            M_7.drop(M_7.index, inplace=True)
            M_7 = data.loc[data.machine == 7].reset_index(drop=True)
        else:
            M_7_Data = M_7[:k7].to_dict(orient = 'record')
            M_7.drop(list(range(0,k7)), inplace=True)
            M_7.reset_index(drop=True, inplace = True)

            
        if len(M_8) < k8:
            if len(M_8) < 10:
                M_8 = data.loc[data.machine == 8].reset_index(drop=True)
            M_8_Data = M_8[:].to_dict(orient = 'record')
            M_8.drop(M_8.index, inplace=True)
            M_8 = data.loc[data.machine == 8].reset_index(drop=True)
        else:
            M_8_Data = M_8[:k8].to_dict(orient = 'record')
            M_8.drop(list(range(0,k8)), inplace=True)
            M_8.reset_index(drop=True, inplace = True)

            
        # Publising data
        publish_message(producer, topic, M_4_Data)
        publish_message(producer, topic, M_5_Data)
        publish_message(producer, topic, M_6_Data)
        publish_message(producer, topic, M_7_Data)
        publish_message(producer, topic, M_8_Data)

        print('publish')
        sleep(5)
```
